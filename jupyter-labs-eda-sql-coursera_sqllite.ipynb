## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
result = engine.execute("SELECT DISTINCT Launch_Site FROM SPACEXTABLE")

# Fetch the launch sites from the query result
launch_sites = [row[0] for row in result.fetchall()]

# Display the launch sites
for site in launch_sites:
    print(site)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
# Execute the query to fetch 5 records where Launch Sites begin with 'CCA'
result = engine.execute("SELECT * FROM SPACEXTABLE WHERE Launch_Site LIKE 'CCA%' LIMIT 5")

# Fetch and display the 5 records
for row in result.fetchall():
    print(row)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
result = engine.execute("SELECT SUM(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Customer = 'NASA (CRS)'")

# Get the total payload mass
total_payload_mass = result.scalar()

# Display the total payload mass carried by boosters launched by NASA (CRS)
print("Total payload mass carried by boosters launched by NASA (CRS):", total_payload_mass, "kg")
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
result = engine.execute("SELECT AVG(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Booster_Version = 'F9 v1.1'")

# Get the average payload mass
average_payload_mass = result.scalar()

# Display the average payload mass carried by booster version F9 v1.1
print("Average payload mass carried by booster version F9 v1.1:", average_payload_mass, "kg")
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
result = engine.execute("SELECT Date FROM SPACEXTABLE WHERE Landing_Outcome = 'Success (ground pad)' ORDER BY Date LIMIT 1")

# Fetch and display the date of the first successful landing on a ground pad
first_successful_landing_date = result.scalar()

print("Date of the first successful landing on a ground pad:", first_successful_landing_date)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
result = engine.execute("SELECT Booster_Version FROM SPACEXTABLE WHERE Landing_Outcome = 'Success (drone ship)' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000")

df = pd.DataFrame.from_dict(result)
df
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
result = engine.execute("SELECT Mission_Outcome, COUNT(*) FROM SPACEXTABLE GROUP BY Mission_Outcome")

df = pd.DataFrame.from_dict(result)
df
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
result = engine.execute("SELECT DISTINCT Booster_Version, MAX(PAYLOAD_MASS__KG_) AS [Maximum Payload Mass] FROM SPACEXTABLE GROUP BY Booster_Version ORDER BY [Maximum Payload Mass] DESC")
    
df = pd.DataFrame.from_dict(result)
df
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
# Execute the adapted query that includes success landing outcomes for the year 2015
query = """
SELECT 
    CASE 
        WHEN SUBSTR(Date, 6, 2) = '01' THEN 'January'
        WHEN SUBSTR(Date, 6, 2) = '02' THEN 'February'
        WHEN SUBSTR(Date, 6, 2) = '03' THEN 'March'
        WHEN SUBSTR(Date, 6, 2) = '04' THEN 'April'
        WHEN SUBSTR(Date, 6, 2) = '05' THEN 'May'
        WHEN SUBSTR(Date, 6, 2) = '06' THEN 'June'
        WHEN SUBSTR(Date, 6, 2) = '07' THEN 'July'
        WHEN SUBSTR(Date, 6, 2) = '08' THEN 'August'
        WHEN SUBSTR(Date, 6, 2) = '09' THEN 'September'
        WHEN SUBSTR(Date, 6, 2) = '10' THEN 'October'
        WHEN SUBSTR(Date, 6, 2) = '11' THEN 'November'
        WHEN SUBSTR(Date, 6, 2) = '12' THEN 'December'
    END AS Month_Name,
    Booster_Version,
    Launch_Site,
    Landing_Outcome
FROM SPACEXTABLE
WHERE SUBSTR(Date, 0, 5) = '2015'
    AND Landing_Outcome LIKE '%Success%'
"""

result = engine.execute(query)

df = pd.DataFrame.from_dict(result)
df
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
query = """
SELECT Landing_Outcome, COUNT(*) AS Outcome_Count
FROM SPACEXTABLE
WHERE Date >= '2010-06-04' AND Date <= '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY Outcome_Count DESC
"""

result = engine.execute(query)

df = pd.DataFrame.from_dict(result)
df
```
