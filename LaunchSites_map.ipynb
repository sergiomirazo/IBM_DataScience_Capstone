<h1 style="text-align: center;"> Launch Sites map</h2>
<hr/>
<p> This interactive map shows SpaceX rocket launchers CCAFS SLC 40, KSC LC 39A and VAFB SLC 4E <p>

```python
#If you dont have folium, then install it using pip3 install folium
import folium

#Launch sites labels:
labels = ["CCAFS SLC 40", "KSC LC 39A", "VAFB SLC 4E"]

#Dictonary with the coordinates in tuples
coor = {
    "CCAFS SLC 40": (28.562301, -80.577356),
    "KSC LC 39A":  (28.573255, -80.646895),
    "VAFB SLC 4E": (34.632834, -120.610746)
}

# define the world map centered around United States of America
usa_map = folium.Map(location=[34.799999,	-92.199997], zoom_start=5)

#loop throught the launch sites coordinates
for label in labels:
    lat = coor[label][0]
    lng = coor[label][1]
    folium.Marker([lat, lng], tooltip='Double click to zoom in', popup=label).add_to(usa_map)
    
# display the USA map
usa_map
```
