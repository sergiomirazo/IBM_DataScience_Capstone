<h1 style="text-align: center;"> Successfully landing outcome map</h2>
<hr/>
<p> This interactive map shows the successfully locations for the SpaceX Falcon9 rockets <p>

```python
import pandas as pd
import folium

# Reading the dataset
df = pd.read_csv("dataset_part_2.csv")

# Filter out rows with 'None None', 'False Ocean' and 'False ASDS' in the Outcome column
filtered_df = df[df['Outcome'] != 'None None']
filtered_df = filtered_df[filtered_df['Outcome'] != 'False Ocean']
filtered_df = filtered_df[filtered_df['Outcome'] != 'False ASDS']

# Extracting Latitude and Longitude columns for the filtered rows
coor = filtered_df[["Latitude", "Longitude"]]

# Convert the filtered data to a dictionary of tuples
coor_dict = {}
keys=[]

for index, row in coor.iterrows():
    #Extract coordinates
    coor_dict[index] = [row["Latitude"], row["Longitude"]]
    keys.append(index)

# Extract columns FlightNumber, Date, BoosterVersion, PayloadMass, Orbit and concatenate the data
filtered_data = filtered_df[["FlightNumber", "Date", "BoosterVersion", "PayloadMass", "Orbit"]]
labels = []

for index, row in filtered_data.iterrows():
    flight_number = row["FlightNumber"]
    date = row["Date"]
    booster = row["BoosterVersion"]
    payload = row["PayloadMass"]
    orbit = row["Orbit"]
    label = f"Flight Number: {flight_number}, Date: {date}, Booster: {booster}, Payload Mass: {payload}, Orbit: {orbit}"
    labels.append(label)
    
    
# define the world map centered around United States of America
usa_map = folium.Map(location=[34.799999,	-92.199997], zoom_start=5)

j = 0
#loop throught the launch sites coordinates
for key in keys:
    lat = coor_dict[key][0]
    lng = coor_dict[key][1]
    label = labels[j]
    folium.Marker([lat, lng], tooltip='Double click to zoom in', popup=label).add_to(usa_map)
    j += 1
    
# display the USA map
usa_map
```
